2. Calculer sa moyenne actuelle et le nombre d'examens restants (2 point)

3. Calculer les moyennes minimales à obtenir pour avoir l'année, ainsi que l'équivalent par note.

````python
# Question 1:

# Déclarer les variables
nb_examen = 15
seuil_valide = 10.0
seuil_felicitation = 12.0
notes = [14.0, 7.5, 19.0, 3.5, 12.0, 16.5, 9.0, 5.5, 18.0, 11.5, 2.0, 15.5]

# Imprimer la nature de données de chaque variable
print(f"La variable 'nb_examen' indique le nombre d'examen à effectuer dans l'année. Sa nature est {str(type(nb_examen))}  : un nombre entier (integer).")
print(f"La variable 'seuil_valide' indique la moyenne minimale pour valider. Sa nature est {str(type(seuil_valide))} : un nombre décimal (float).")
print(f"La variable 'seuil_felicitation' indique la moyenne minimale pour avoir les Félicitations. Sa nature est {str(type(seuil_felicitation))} : un nombre décimal (float).")
print(f"La variable 'notes' contient les notes des examens déjà effectués. Sa nature est {str(type(notes))} : une liste.")
````

````python
# Question 2

# Calculer la sum des notes
somme_notes_actuelle = sum(notes)

# Obtenir le nombre d'examens déjà effectués
nb_exam_effectue = len(notes)

# Calculer la moyenne
moyenne = somme_notes_actuelle/nb_exam_effectue
print(f"La moyenne actuelle est {moyenne}.")

# Calculer le nombre d'examens restant
nb_exam_restant = nb_examen - nb_exam_effectue
print(f"Il reste {nb_exam_restant} examens.")
````

````python
# Question 3:

# Calculer la somme minimale des notes pour valider et pour avoir des Félicitations
somme_minimal_valide = nb_examen*seuil_valide
somme_minimal_felicitation = nb_examen*seuil_felicitation

# Calculer la moyenne minimale pour les notes d'examens restant afin de valider, ainsi que l'équivalent en notes
somme_min_restant = somme_minimal_valide - somme_notes_actuelle
moyenne_min_valide = somme_min_restant/nb_exam_restant

print(f"La moyenne minimale à obtenir pour avoir l'année est {moyenne_min_valide}, l'équivalent par note est {somme_min_restant}.")

# Calculer la moyenne minimale pour les notes d'examens restant afin d'avoir les félicitations, ainsi que l'équivalent en notes
somme_restant_felicitation = somme_minimal_felicitation - somme_notes_actuelle
moyenne_min_felicitation = somme_restant_felicitation/nb_exam_restant

if moyenne_min_felicitation > 20:
  print(f"C'est impossible d'avoir les félicitations. Il faut avoir une moyenne de {moyenne_min_felicitation}, supérieur à 20, sur {nb_exam_restant} examens.")
else : print(f"La moyenne minimale à obtenir pour avoir les félicitations est {moyenne_min_felicitation}, l'équivalent par note est {somme_restant_felicitation}.")




````

````python
# Question 4

# Déclaration
notes_avec_coef = {
    "note_1": {"note": 14.0, "coef": 1},
    "note_2": {"note": 7.5,  "coef": 3},
    "note_3": {"note": 19.0, "coef": 1},
    "note_4": {"note": 3.5,  "coef": 2},
    "note_5": {"note": 12.0, "coef": 2},
    "note_6": {"note": 16.5, "coef": 1},
    "note_7": {"note": 9.0,  "coef": 2},
    "note_8": {"note": 5.5,  "coef": 3},
    "note_9": {"note": 18.0, "coef": 1},
    "note_10": {"note": 11.5, "coef": 1},
    "note_11": {"note": 2.0,  "coef": 2},
    "note_12": {"note": 15.5, "coef": 1},
}

# éditer la note:
notes_avec_coef["note_9"] = {"note": 14.0, "coef": 1}
print(f"La 9ème note est corrigé à {notes_avec_coef["note_9"]}.")
````

````python
# Question 5

# Déclaration
somme_note_coeff = 0
somme_coeff = 0

# Boucle for pour parcourir le dictionnaire
for note in notes_avec_coef:
  somme_note_coeff = somme_note_coeff + notes_avec_coef[note]["note"]*notes_avec_coef[note]["coef"]
  somme_coeff = somme_coeff + notes_avec_coef[note]["coef"]

moyenne_coef = somme_note_coeff/somme_coeff
print(f"La moyenne actuelle est {moyenne_coef}.")
print(somme_note_coeff)
print(somme_coeff)

# Calculer la moyenne minimale pour les notes d'examens restant afin de valider, ainsi que l'équivalent en notes
somme_minimal_valide = moyenne_min_valide*(somme_coeff+3)
somme_min_restant = moyenne_min_valide*(somme_coeff+3) - moyenne_coef*somme_coeff
moyenne_min_valide = somme_min_restant/nb_exam_restant

print(f"La moyenne minimale à obtenir pour avoir l'année est {moyenne_min_valide}, l'équivalent par note est {somme_min_restant}.")

# Calculer la moyenne minimale pour les notes d'examens restant afin d'avoir les félicitations, ainsi que l'équivalent en notes
somme_minimal_felicitation = moyenne_min_felicitation*(somme_coeff+3)
somme_restant_felicitation = moyenne_min_felicitation*(somme_coeff+3) - moyenne_coef*somme_coeff
moyenne_min_felicitation = somme_restant_felicitation/nb_exam_restant

if moyenne_min_felicitation > 20:
  print(f"C'est impossible d'avoir les félicitations. Il faut avoir une moyenne de {moyenne_min_felicitation}, supérieur à 20, sur {nb_exam_restant} examens.")
else : print(f"La moyenne minimale à obtenir pour avoir les félicitations est {moyenne_min_felicitation}, l'équivalent par note est {somme_restant_felicitation}.")
````
