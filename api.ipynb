```python
import requests
```

```python
# l'endpoint
url = "https://api.open-meteo.com/v1/forecast"

# Les paramètres (query parameters)
params = {
    "latitude": 50.633,
    "longitude": 3.0586,
    "hourly": "temperature_2m,rain,wind_speed_10m",
    "current": "temperature_2m,wind_speed_10m",
    "timezone": "Europe/Paris",
}

# La réponse
resp = requests.get(url, params=params, timeout=20)
```

```python
# Afficher l'ensemble de la réponse
data = resp.json()
data
```

**API Open AI**

```python
# Des imports
from openai import OpenAI
import os
from getpass import getpass
```

```python
# Pour la clé API
os.environ["OPENAI_API_KEY"] = getpass("Colle la clé de l'intervenant (elle restera cachée) : ")
```

```python
client = OpenAI()  # lit OPENAI_API_KEY depuis l'environnement

# lance la requête
response = client.responses.create(
    model="gpt-4.1-mini",
    input="Donne moi une recette de tarte aux pruneaux."
)

print(response.output_text)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input=[
        {"role": "system", "content": "Tu es un assistant utile et concis."},
        {"role": "user", "content": "Donne-moi une recette de tarte aux pruneaux."}
    ]
)

print(response.output_text)
```
