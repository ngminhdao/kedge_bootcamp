**Exercice 2/2 :**

- Calcule la moyenne de Salma, Morgane et Anne.

- Ajoute ton nom avec 5 notes dans la liste.

- Retire la première clé-valeur "Léa" de la liste.

Indices  

- la méthode sum() applicable à une list

```python
import random as r

notes = {
    "Léa":   [14, 16, 13, 15, 17],
    "Hugo":  [10, 12,  9, 11, 13],
    "Salma": [18, 17, 19, 16, 18],
    "Morgane":[12, 14, 13, 15, 11],
    "Sasha": [15, 16, 14, 13, 17],
    "Anne":  [ 9, 10, 11, 12,  8],
    "Yasmine":[16, 15, 17, 14, 18],
}

#liste qui contient les noms/clés à calculer les moyennes
q1 = ["Salma", "Morgane", "Anne"]

#boucle pour calculer et imprimer la moyenne
for i in q1:
  print(f"La moyenne de {i} est {sum(notes.get(i))/len(notes.get(i))}.")

#liste vide pour randomiser les notes
q2 = []

#boucle for pour faire 5 notes
for i in range(5):
  q2.append(r.randint(8,19))

#déclare ton nom
prenom = "Minh-Dao"

#update notes dictionnaire avec ton nom
notes.update(
    {prenom:q2}
)

#print ta moyenne
print(f"La moyenne de {prenom} est {sum(notes.get(prenom))/5}.")

```

**Exercice 2/3**

Compter les occurrences du mot 'python' dans la phrase :

phrase = "python python boucle dictionnaire python boucle"

Indices :

Découpe la phrase en mots grâce à la méthode .split()

Avec une boucle, créé une variable qui comptabilise les occurrences

```python
#déclarer la phrase
phrase = "python python boucle dictionnaire python boucle"
print(phrase)

#déclarer une liste qui contient les mots après avoir splitté la phrase
mots = phrase.split()

#print liste de mots
print(mots)

#déclarer compteur
compteur = 0

#boucle for qui parcourut la liste, compte "python"
for i in mots:
  if i == "python": compteur += 1

print(f"Dans la phrase donnée, il y a {compteur} occurrences du mot 'python'")
```
